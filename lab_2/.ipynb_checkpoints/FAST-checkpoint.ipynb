```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('img_1.png',  cv2.IMREAD_COLOR)

plt.imshow(img)
plt.show()
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # this converts it into RGB

plt.imshow(rgb_img)
plt.show()
```

```python
# Initiate FAST object with default values
fast = cv2.FastFeatureDetector_create()

kp = fast.detect(rgb_img, None)
```

```python
br = cv2.BRISK_create();
kp1, des1 = br.compute(rgb_img,  kp)
```

```python
test_image = rgb_img.copy()
num_rows, num_cols = test_image.shape[:2]

rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), -25, 1)
test_image = cv2.warpAffine(test_image, rotation_matrix, (num_cols, num_rows))

test_image = test_image[165:210, 35:70]



plt.imshow(test_image)
plt.show()
```

```python
kp2 = fast.detect(test_image, None)
kp2, des2 = br.compute(test_image,  kp2)
```

```python
des1.shape
```

```python
br.compute(test_image,  kp2)
```

```python
# create BFMatcher object 
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True) 
# Match descriptors. 
matches = bf.match(des1, des2) 
# Sort them in the order of their distance. 
matches = sorted(matches, key = lambda x:x.distance) 
# Draw first 10 matches. 
img3 = cv2.drawMatches(rgb_img,kp1,test_image,kp2,matches[:10],None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS) 
plt.figure(figsize=(10, 7))
plt.imshow(img3)
plt.show() 
```

```python
def matcher(des1: np.ndarray, des2: np.ndarray, threshold, distf) -> list:
    matches = []
    for i for range(len(des1)):
        for j in range(len(des2)):
            dst = distf(desc1[i].reshape(1,-1), desc2[j].reshape(1,-1))
            if dst < 60:
                matches.append(cv2.DMatch(_queryIdx=i, _trainIdx=j, _imgIdx=0,_distance=dst))
    
    return matches
```

```python
class LinearLeastSqaureModel:
    def fit(self, A, Y):
        A_transpose = A.transpose()
        ATA = A_transpose.dot(A)
        ATY = A_transpose.dot(Y)
        model = (np.linalg.inv(ATA)).dot(ATY) ## For a linear eq. AP = Y to solve a least sqaure problem,  P = (inverse(A'A))(A'Y) 
        return model
```
